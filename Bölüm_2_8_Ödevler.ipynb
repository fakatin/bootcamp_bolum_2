```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1,2,1)
x=pd.read_csv('ATABEY2.csv')
df=pd.DataFrame(x)
y_pos=np.arange(len(df['BOLUM']))
plt.bar(y_pos,df['GNO_ORT'],width=0.5,color='yellow')
plt.xticks(y_pos, df['BOLUM'])
plt.ylabel('GNO')
plt.xlabel('Bölümler')
plt.title('BÖLÜM GNO ORTALAMALARI')

plt.subplot(1,2,2)
x=pd.read_csv('ATABEY2.csv')
df=pd.DataFrame(x)
y_pos=np.arange(len(df['BOLUM']))
plt.bar(y_pos,df['YP_ORT'],width=0.5,color='red')
plt.xticks(y_pos, df['BOLUM'])
plt.ylabel('YP')
plt.xlabel('Bölümler')
plt.title('BÖLÜM YERLEŞTRME PUAN ORTALAMALARI')


plt.show()
```
